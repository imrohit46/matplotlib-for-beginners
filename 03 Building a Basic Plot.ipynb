```python
plt.plot([6,8,4,2,3,6,8,4,5,2])
plt.show()
```

**Changing Marker Type:**

```python
plt.plot([6,8,4,2,3,6,8,4,5,2], 'o')
plt.show()
```

```python
plt.plot([6,8,4,2,3,6,8,4,5,2], '^')
plt.show()
```

**Changing Color and Line Style:**

```python
plt.plot([6,8,4,2,3,6,8,4,5,2], marker="o", color="green", linestyle="dashed")
plt.show()
```

**Marker and Color in the same argument:**

```python
plt.plot([6,8,4,2,3,6,8,4,5,2], "go")
plt.show()
```

```python
plt.plot([6,8,4,2,3,6,8,4,5,2], "k^")
plt.show()
```
