```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 2x2 Grid

plt.figure(figsize=(12,8))

plt.subplot(221)
plt.title('Index Position 1')
plt.annotate(text='1', xy=(0.5, 0.5), fontsize=45)

plt.subplot(222)
plt.title('Index Position 2')
plt.annotate(text='2', xy=(0.5, 0.5), fontsize=45)

plt.subplot(223)
plt.title('Index Position 3')
plt.annotate(text='3', xy=(0.5, 0.5), fontsize=45)

plt.subplot(224)
plt.title('Index Position 4')
plt.annotate(text='4', xy=(0.5, 0.5), fontsize=45)
```

```python
data = pd.read_csv("iris.csv")
sl = data["sepal_length"]
sw = data["sepal_width"]
pl = data["petal_length"]
pw = data["petal_width"]
species = data["species"]

data2 = data.groupby('species', as_index=False).mean()
```

```python
stock = pd.read_csv("stock.csv")
stock['timestamp'] = pd.to_datetime(stock['timestamp'])

dates = stock['timestamp']
volume = stock['volume']
price = stock['close']
```

```python
# 2x2 grid filled with data visualization

plt.figure(figsize=(12,8))

plt.subplot(221)
plt.scatter(sl, sw, marker='s', color='red')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Sepal Width')

plt.subplot(222)
plt.plot(sl, sw)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Sepal Width')

plt.subplot(223)
plt.bar(data2['species'], data2['sepal_length'], width=0.5)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Sepal Width')

plt.subplot(224)
plt.plot(dates, volume)
plt.xlabel("Date")
plt.ylabel("Volume")
plt.title("Volume Traded on the Date")

plt.tight_layout()

plt.show()
```

```python
# Layout 2x1

plt.figure(figsize=(12,8))

plt.subplot(211)
plt.annotate(text='1', xy=(0.4, 0.4), fontsize=45)

plt.subplot(212)
plt.annotate(text='2', xy=(0.4, 0.4), fontsize=45)

plt.tight_layout()

plt.show()
```
