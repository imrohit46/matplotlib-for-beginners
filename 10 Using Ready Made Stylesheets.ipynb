```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("iris.csv")
sl = data["sepal_length"]
sw = data["sepal_width"]
pl = data["petal_length"]
pw = data["petal_width"]
species = data["species"]
```

**Custom Formatting:**

Custom formats add extra lines to the code which is pententially repetitive.

```python
plt.figure(figsize=(8,4), facecolor='yellow')
plt.scatter(sl, sw, edgecolors='black', facecolor='red', linewidths=2, s=100)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.tick_params(axis='both', labelsize='large', length=10, color='green')
plt.title('Sepal Length vs Sepal Width', fontsize=24, loc='left', fontstyle='italic', color='blue')
plt.grid(color='gray', linestyle='--', linewidth=1)
plt.show()
```

**Style Sheet:**

A container of code for formatting preferences. It can be applied to one or more data visualizations with just a single call.

```python
plt.style.available
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(12,6))
plt.scatter(sl, sw)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```
