**Main argument types for coloration:**
1. Color - For generic colors and font colors.
2. Edge Color - 'edgecolor'
3. Face Color - 'facecolor'

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("iris.csv")
sl = data["sepal_length"]
sw = data["sepal_width"]
pl = data["petal_length"]
pw = data["petal_width"]
species = data["species"]
```

```python
# Background color of the figure
plt.figure(figsize=(12,6), facecolor='yellow')
plt.scatter(sl, sw)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
# Formatting the markers
plt.figure(figsize=(12,6), facecolor='yellow')
plt.scatter(sl, sw, edgecolors='black')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
#size of markers and line width
plt.figure(figsize=(12,6), facecolor='yellow')
plt.scatter(sl, sw, edgecolors='black', facecolor='red', linewidths=2, s=100)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
# Format text objects: Axis labels & Tickers
plt.figure(figsize=(8,4), facecolor='yellow')
plt.scatter(sl, sw, edgecolors='black', facecolor='red', linewidths=2, s=100)
plt.xlabel('Sepal Length', fontsize=20)
plt.ylabel('Sepal Width', fontsize=20)
plt.tick_params(axis='both', labelsize='large', length=10, color='green')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
#Format text objects: Main Title
plt.figure(figsize=(8,4), facecolor='yellow')
plt.scatter(sl, sw, edgecolors='black', facecolor='red', linewidths=2, s=100)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.tick_params(axis='both', labelsize='large', length=10, color='green')
plt.title('Sepal Length vs Sepal Width', fontsize=24, loc='left', fontstyle='italic')
plt.show()
```

```python
# Additional grid line for better readability
plt.figure(figsize=(8,4), facecolor='yellow')
plt.scatter(sl, sw, edgecolors='black', facecolor='red', linewidths=2, s=100)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.tick_params(axis='both', labelsize='large', length=10, color='green')
plt.title('Sepal Length vs Sepal Width', fontsize=24, loc='left', fontstyle='italic', color='blue')
plt.grid(color='gray')
plt.show()
```

```python
# Line style of the Grid
plt.figure(figsize=(8,4), facecolor='yellow')
plt.scatter(sl, sw, edgecolors='black', facecolor='red', linewidths=2, s=100)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.tick_params(axis='both', labelsize='large', length=10, color='green')
plt.title('Sepal Length vs Sepal Width', fontsize=24, loc='left', fontstyle='italic', color='blue')
plt.grid(color='gray', linestyle='--', linewidth=5)
plt.show()
```
