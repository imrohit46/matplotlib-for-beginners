```python
import numpy as np
import pandas as pd
```

**Reading the Dataset:**

```python
data = pd.read_csv("iris.csv")
data.head()
```

**Plotting Using Pandas Integrated Matplotlib:**

```python
data.plot(kind='scatter', x='sepal_length', y='sepal_width')
```

```python
# adding markers, colors and titles
data.plot(kind='scatter', x='sepal_length', y='sepal_width', marker='s', color='green', title='Sepal Length vs Sepal Width')
```

```python
import matplotlib.pyplot as plt
```

* Mixing of Pyplot and Pylab tools is possible but not recommended due to the risk of bugs and inconsistencies in the code.

```python
data.plot(kind='scatter', x='sepal_length', y='sepal_width', marker='s', figsize=(12,6))
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
sl = data["sepal_length"]
sw = data["sepal_width"]
```

**Standard Matplotlib Code (Pyplot):**

```python
plt.figure(figsize=(12,6))
plt.scatter(sl, sw, marker='s', color='green')
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```
